# Deep Learning

## ANN Regression
Artificial Neural Network Regression model to predict the electrical energy output of a Combined Cycle Power Plant

#### Implement the libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

### Part 1 - Data Preprocessing

#### Import the dataset

```python
dataset = pd.read_excel('Folds5x2_pp.xlsx', engine='openpyxl')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

#### Split dataset into training and test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Part 2 - Building the ANN

#### Initializing the ANN

```python
ann = tf.keras.models.Sequential()
```

#### Adding the Input Layer and the First Hidden Layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

#### Adding the Second Hidden Layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

#### Adding the Output Layer

```python
ann.add(tf.keras.layers.Dense(units=1))
```

### Part 3 - Training the ANN

#### Compiling the ANN

```python
ann.compile(optimizer='adam', loss='mean_squared_error')
```

#### Training the ANN on the training set

```python
ann.fit(x=X_train, y=y_train, epochs=100, batch_size=32)
```

#### Predict the results of the Test set

```python
y_pred = ann.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```
